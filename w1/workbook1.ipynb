```python
import numpy as np
import matplotlib.pyplot as plt
# from scipy.integrate import simps

import math

def some_func(x):
    return math.sqrt(1 + math.e ** math.sqrt(x) + math.cos(x**2)) / abs(1 - math.sin(x) ** 3) + math.log(abs(2 * x))

x = np.arange(1,10,0.1)
y = []
for i in x:
    y.append(some_func(i))
plt.grid()
plt.plot(x, y, c = "r")
plt.fill_between(x, y)
plt.show()

fig, ax = plt.subplots()
ax.scatter(x = x[len(x)//2:], y = y[len(x)//2:])

plt.show()
```

```python
import numpy as np
operation = input('''
Please type in the math operation you would like to complete:
+ for addition
- for subtraction
* for multiplication
/ for division
e for e^(x+y)
sin for sin(x+y)
cos for cos(x+y)
x^y for x^y
''')
x = float(input("Please type x"))
y = float(input("Please type y"))
if operation == "+":
    print(x+y)
elif operation == "-":
    print(x-y)
elif operation == "*":
    print(x*y)
elif operation == "/":
    print(x/y)
elif operation == "e":
    print(np.exp(x+y))
elif operation == "sin":
    print(np.sin(x+y))
elif operation == "cos":
    print(np.cos(x+y))
elif operation == "x^y":
    print(x**y)
else:
    print("Wrong operation")
```
